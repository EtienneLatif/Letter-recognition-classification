# Activation functions

## Sigmoid function

```python
# Sigmoid function
def sigmoid(z):
    return 1 / (1+np.exp(-z))

# First derivative of sigmoid function with respect to z
def der_sigmoid(z):
    return sigmoid(z)*(1-sigmoid(z))
```

```python
z = np.linspace(-10, 10, 2000) # Generate data points
sig_z = sigmoid(z) # Apply sigmoid function
der_sig_z = der_sigmoid(z) # First derivative of sigmoid function
# Plotting the sigmoid function
%matplotlib inline
fig, ax = plt.subplots(1,1, figsize=(15, 10))
ax.plot(z, sig_z, label='Sigmoid function')
ax.plot(z, der_sig_z, label='First derivative of sigmoid function')
ax.set_xlim(-10, 10)
ax.axhline(y=0, color='k', alpha=0.25)
ax.axhline(y=1, color='k', alpha=0.25, linestyle='--')
ax.axvline(x=0, color='k', alpha=0.25)
ax.legend(fontsize=17)
```

## ReLU and PReLU functions

```python
# ReLU function
def relu(z):
    return np.max([0, z])

# First derivative of ReLU function with respect to z
def der_relu(z):
    if z > 0: return 1
    else: return 0
    
# PReLU function
def prelu(a,z):
    if z > 0: return z
    else: return a*z
    
# First derivative of PReLU function with repsect to z
def der_prelu(a, z):
    if z > 0: return 1
    else: return a
```

```python
z = np.linspace(-5, 5, 2000) # Generate data points
relu_z = [relu(z) for z in z] # Apply ReLU function
der_relu_z = [der_relu(z) for z in z] # First derivative of ReLU
%matplotlib inline
fig, ax = plt.subplots(1,1, figsize=(15, 10))
ax.plot(z, relu_z, label='ReLU function')
ax.plot(z, der_relu_z, label='First derivative of ReLU function')
ax.set_xlim(-5, 5)
ax.axhline(y=0, color='k', alpha=0.25)
ax.axvline(x=0, color='k', alpha=0.25)
ax.legend(fontsize=17)
```

```python
z = np.linspace(-10, 10, 2000) # Generate data points
prelu_z = [prelu(0.1, z) for z in z] # Apply ReLU function
der_prelu_z  = [der_prelu(0.1, z) for z in z]
# First derivative of ReLU
%matplotlib inline
fig, ax = plt.subplots(1,1, figsize=(15, 10))
ax.plot(z, prelu_z, label='PReLU function, a=0.1')
ax.plot(z, der_prelu_z, label='First derivative of PReLU function')
ax.set_xlim(-3, 3)
ax.set_ylim(-1, 2)
ax.axhline(y=0, color='k', alpha=0.25)
ax.axvline(x=0, color='k', alpha=0.25)
ax.legend(fontsize=17)
```
